```python
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
paragraph = """Artificial intelligence is becoming one of the most important technologies in the world because it is helping people, businesses, students, doctors,
engineers, and researchers solve many different kinds of problems in a faster and smarter way. In the past, most computer programs were written with fixed rules, and they
could only do exactly what the programmer told them to do, but today many AI systems can learn from data, understand patterns, make predictions, answer questions,
summarize documents, translate languages, and even generate human-like text. For example, when a student uploads a document into an AI study assistant, the system can read
the material, remove unnecessary words, identify important concepts, and generate useful answers from the content. However, before a machine learning or natural language
processing model can understand text properly, the text usually needs to be cleaned and prepared. This process may include converting the text into lowercase, removing
punctuation, splitting the sentence into tokens, removing stop words, applying stemming or lemmatization, and finally converting the words into numerical features.
Stop words are common words such as “the,” “is,” “and,” “in,” “to,” “of,” “for,” “with,” “a,” and “an” that often appear many times in a paragraph but may not always
add strong meaning to the main idea. When these words are removed, the remaining words can give a clearer view of the important topics in the text. For example, in a
sentence about building, deploying, and optimizing GenAI applications with LangChain, words like “building,” “deploying,” “optimizing,” “GenAI,” “applications,”
and “LangChain” are more meaningful than small connecting words like “and,” “with,” or “the.” This is why stop words removal is useful in many NLP tasks such as
search engines, sentiment analysis, document classification, chatbot development, recommendation systems, and text summarization."""
```

```python
ps  = PorterStemmer()
ss = SnowballStemmer(language='english')
lm = WordNetLemmatizer()
```

```python
sentences = nltk.sent_tokenize(paragraph)
```

```python
## Apply stopwords and filter and then apply snowball stemming

for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [ss.stem(word) for word in words if word not in set(stopwords.words('english'))]
  sentences[i] = ' '.join(words) ## coverting all the list of words into sentences

```

```python
sentences
```

```python
for i in range(len(sentences)):

  words = nltk.word_tokenize(sentences[i])
  words = [lm.lemmatize(word.lower(),pos='v') for word in words if word not in stopwords.words('english')]
  sentences[i] = ' '.join(words)
```
