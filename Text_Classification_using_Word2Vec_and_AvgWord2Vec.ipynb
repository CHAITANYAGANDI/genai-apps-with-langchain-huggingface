```python
import pandas as pd

messages = pd.read_csv('spam.csv',encoding='latin-1')

messages = messages[['v1','v2']]
messages.columns = ['label','text']
messages.head()
```

```python
import nltk
import re
import gensim
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import sent_tokenize
from tqdm import tqdm
import numpy as np
from gensim.utils import simple_preprocess
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
lm = WordNetLemmatizer()
```

```python
corpus = []

for i in range(0,len(messages)):
  review = re.sub('[^a-zA-Z]',' ',messages['text'][i])
  review = review.lower()
  review = review.split()
  review = [lm.lemmatize(word) for word in review if word not in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
words = []

for sent in corpus:
  sent_token = sent_tokenize(sent)
  for word in sent_token:
    words.append(simple_preprocess(word))
```

```python
words
```

```python
model = gensim.models.Word2Vec(words)
```

```python
## To get all the vocabulary
model.wv.index_to_key
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
model.wv.similar_by_word("good")
```

```python
model.wv['good']
```

```python
def avg_word2vec(doc):
  # axis=0 ensures we average the vectors, not the individual scalars
  # We also handle the case where a document might be empty to avoid errors
  vecs = [model.wv[word] for word in doc if word in model.wv.index_to_key]
  if not vecs:
    return np.zeros(model.vector_size)
  return np.mean(vecs, axis=0)
```

```python
X = []

for i in tqdm(range(len(words))):
  X.append(avg_word2vec(words[i]))
```

```python
len(X)
```

```python
## Dependent features
y = messages[list(map(lambda x: len(x)>0, corpus))]
y = pd.get_dummies(y['label'])
y = y.iloc[:,0].values
```

```python
y.shape
```

```python
y.shape
```

```python
df = pd.DataFrame()

for i in range(0, len(X)):
    row = pd.DataFrame(X[i].reshape(1, -1))
    df = pd.concat([df, row], ignore_index=True)
```

```python
df.shape
```

```python
X = df
```

```python
from sklearn.model_selection import train_test_split

# Ensure X and y are correctly aligned and split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and fit the model
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
```

```python
rf_model.score(X_test, y_test)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
