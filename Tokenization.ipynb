```python
corpus = """NorthLedger is a personal finance app that helps users
track income, expenses, transfers, and budgets. On July 6, 2026,
Chaitanya added a $45.75 grocery transaction at Walmart, marked it as a
debit, and linked it to the “Food” category. Later, he transferred $200
from his chequing account to his savings account. The app should
correctly split words, numbers, dates, punctuation, symbols, and quoted
text into tokens.
"""
```

```python
print(corpus)
```

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize
```

```python
docs = sent_tokenize(corpus)
```

```python
type(docs)
```

```python
for doc in docs:
  print(doc)
```

```python
from nltk.tokenize import word_tokenize
```

```python
words = word_tokenize(corpus)
```

```python
for sentences in docs:
  print(word_tokenize(sentences))
```

```python
for sentences in docs:
  print(nltk.wordpunct_tokenize(sentences))
```

```python
for sentences in docs:
  print(nltk.TreebankWordTokenizer().tokenize(sentences))
```
