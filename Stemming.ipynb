```python
## Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or
## to the roots of words known as lemma. Stemming is important in natural language understandin and processing.

## Classificaion Problem
## comments of product is a positive review or negative review
## Reviews -----> filter out positive one and negative ones

## [eating, eat, eaten] - final meaning is eat
## [going, gone, goes] - final meaning is go


words = [
    "eating",
    "eat",
    "eaten",
    "writing",
    "writes",
    "programming",
    "programs",
    "history",
    "finally",
    "finalized"
]
```

```python
## First stemming technique is Porter Stemmer

from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
for word in words:

  print(word+"---->"+ps.stem(word))
```

```python
## This is the major disadvantage of the stemming because for some word it is not extracting exact meaning
## To overcome this we have to use lemmatization
ps.stem("Congratulations")
```

```python
## Second stemming technique is RegexpStemmer Class

from nltk.stem import RegexpStemmer
rs = RegexpStemmer("ing$|e$|ed$|s$|able$",min=4)
```

```python
rs.stem('eating')
```

```python
## Third stemming technique is Snowball Stemmer -
##this is best stemmer out of the above two stemming techniques

from nltk.stem import SnowballStemmer
ss = SnowballStemmer("english")
```

```python
for word in words:

  print(word+"---->"+ss.stem(word))
```

```python
ps.stem("fairly"), ps.stem("sportingly")
```

```python
ss.stem("fairly"), ss.stem("sportingly")
```

```python
## for the majority of the words will get correct word meaning through snowball stemming but not all most all word doesn't get accurate results
## so we have to go through lemmatization for that because it has all dictionary of words - if you give any word it will give good form of grammitcal exact meaning word.
```
