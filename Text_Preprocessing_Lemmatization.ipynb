# **Wordnet Lemmatizer**

Lemmatization technique is like stemming. The output we will get after lemmatization is called 'lemma'. which is a root word rather than root stem, the output of stemming. After lemmatization, we will be getting a valid word that means the same thing.

NLTK provides WordNetLemmatizer class which is a thin wrapper around the wordnet corpus. This class uses morphy() function to the WordNet CorpusReader class to find a lemma.

Usecases: Q&A, Chatbots, Text Summarization.

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
print(lemmatizer.lemmatize("cacti"))
print(lemmatizer.lemmatize("mice"))
print(lemmatizer.lemmatize("rocks"))
```

```python
## there is one parameter called pos that we have to pass while doing the lemmatization

'''
POS - Parts of speech

Noun - n, Verb - v, Adjective - a, Adverb - r
'''

print(lemmatizer.lemmatize("better", pos="a"))
print(lemmatizer.lemmatize("best", pos="a"))
```

```python
print(lemmatizer.lemmatize("am", pos="v"))
print(lemmatizer.lemmatize("are", pos="v"))
print(lemmatizer.lemmatize("is", pos="v"))
```

```python
print(lemmatizer.lemmatize("going",pos='v'))
```

```python
words = [
    "eating",
    "eat",
    "eaten",
    "writing",
    "writes",
    "programming",
    "programs",
    "history",
    "finally",
    "finalized"
]
```

```python
for word in words:
  print(lemmatizer.lemmatize(word, pos="v"))
```
